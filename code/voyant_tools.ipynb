```python
from pathlib import Path

# ── ПОМЕНЯЙ НА СВОЙ ПУТЬ ──────────────────────────────────────────
BASE_PATH = Path('/content/drive/MyDrive/SFU 4/VKR/Processed')
```

```python
# В Colab — создаём выборки по 500 файлов из каждого класса для Voyant
import random

for label in ['depression', 'anxiety', 'neutral']:
    files = list((BASE_PATH / 'labeled' / label).glob('*.txt'))
    sample = random.sample(files, min(500, len(files)))

    # Объединяем в один файл — Voyant хорошо работает с большими txt
    out_text = '\n\n'.join([f.read_text(encoding='utf-8') for f in sample])

    out_path = f'/content/drive/MyDrive/SFU 4/VKR/Voyant Tools/voyant_{label}.txt'
    with open(out_path, 'w', encoding='utf-8') as f:
        f.write(out_text)
    print(f'{label}: {len(sample)} файлов → voyant_{label}.txt')
```
