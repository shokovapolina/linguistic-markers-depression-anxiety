```python
import os
import re
import nltk
from nltk.corpus import stopwords
!pip install pymorphy3
from pymorphy3 import MorphAnalyzer
```

```python
# Скачиваем стоп-слова для русского языка
nltk.download('stopwords')
stop_words = set(stopwords.words('russian'))

# Инициализация анализатора
morph = MorphAnalyzer()

def clean_text(text, remove_stopwords=True):
#Функция для предварительной очистки текста (предобработка)
    text = re.sub(r'[^а-яА-ЯёЁ\s]', ' ', text)  # Удаляем всё, кроме букв и пробелов
    text = re.sub(r'\s+', ' ', text)  # Удаляем лишние пробелы
    text = text.lower().strip()  # Приводим к нижнему регистру и обрезаем пробелы

    if remove_stopwords:
      words = text.split()
      words = [word for word in words if word not in stop_words]
      text = ' '.join(words)
    return text

def lemmatize_text(text, remove_stopwords=True):
#Функция для лемматизации текста
    words = text.split()
    lemmatized = []
    for word in words:
        if remove_stopwords and word in stop_words:
            continue
        lemma = morph.parse(word)[0].normal_form
        lemmatized.append(lemma)
    return ' '.join(lemmatized)

def process_file(file_path, input_root, output_root):
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()

    # Предобработка
    text_clean = clean_text(text)
    text_lemma = lemmatize_text(text_clean)

    # Определяем относительный путь
    relative_path = os.path.relpath(file_path, input_root)
    relative_folder = os.path.dirname(relative_path)
    base_name = os.path.basename(file_path)

    # Добавляем префиксы
    clean_filename = f"clean_{base_name}"
    lemma_filename = f"lemma_{base_name}"

    # Папки Clean и Lemma
    clean_folder = os.path.join(output_root, 'Clean', relative_folder)
    lemma_folder = os.path.join(output_root, 'Lemma', relative_folder)

    os.makedirs(clean_folder, exist_ok=True)
    os.makedirs(lemma_folder, exist_ok=True)

    # Сохраняем
    with open(os.path.join(clean_folder, clean_filename), 'w', encoding='utf-8') as f:
        f.write(text_clean)

    with open(os.path.join(lemma_folder, lemma_filename), 'w', encoding='utf-8') as f:
        f.write(text_lemma)

    print(f"Обработан: {relative_path}")


def main():
    input_root = '/content/drive/My Drive/SFU 4/VKR/csvs/TXT_OUTPUT'
    output_root = '/content/drive/My Drive/SFU 4/VKR/Processed/'

    for foldername in os.listdir(input_root):
        folder_path = os.path.join(input_root, foldername)

        if os.path.isdir(folder_path):  # Проверяем, что это папка
            for filename in os.listdir(folder_path):
                if filename.endswith('.txt'):
                    file_path = os.path.join(folder_path, filename)
                    process_file(file_path, input_root, output_root)

if __name__ == '__main__':
    main()
```

# **Объедененные файлы**

```python
def process_single_file(file_path, output_folder):
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()

    text_clean = clean_text(text)
    text_lemma = lemmatize_text(text_clean)

    base_name = os.path.basename(file_path)

    clean_filename = f"clean_{base_name}"
    lemma_filename = f"lemma_{base_name}"

    os.makedirs(output_folder, exist_ok=True)

    with open(os.path.join(output_folder, clean_filename), 'w', encoding='utf-8') as f:
        f.write(text_clean)

    with open(os.path.join(output_folder, lemma_filename), 'w', encoding='utf-8') as f:
        f.write(text_lemma)

    print("Файл обработан ✅")


# ===== запуск =====
file_path = '/content/drive/My Drive/SFU 4/VKR/Processed/combined_texts/Все тексты.txt'
output_folder = '/content/drive/My Drive/SFU 4/VKR/Processed/combined_texts/'

process_single_file(file_path, output_folder)
```
